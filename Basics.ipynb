### Task#8

```python

# Global variables to store the selected vertices and the image
vertices = []
img = None

def on_mouse_click(event, x, y, flags, param):
    global vertices, img
    
    if event == cv2.EVENT_LBUTTONDOWN:
        # Add point to vertices
        vertices.append((x, y))
        # Draw a small circle at the clicked point
        cv2.circle(img, (x, y), 3, (0, 255, 0), -1)
        # Draw lines between points
        if len(vertices) > 1:
            cv2.line(img, vertices[-2], vertices[-1], (0, 255, 0), 2)
        cv2.imshow('Image', img)

    elif event == cv2.EVENT_RBUTTONDOWN:
        # Close the polygon
        if len(vertices) > 2:
            cv2.line(img, vertices[-1], vertices[0], (0, 255, 0), 2)
            # Convert vertices list to numpy array
            pts = np.array(vertices, np.int32)
            pts = pts.reshape((-1, 1, 2))
            # Draw the filled polygon
            cv2.fillPoly(img, [pts], (255, 0, 0))
            cv2.imshow('Image', img)
            # Print the vertices
            print("Vertices of the selected polygon:")
            for vertex in vertices:
                print(vertex)

def apply_translation(points, tx, ty):
    #Apply translation to the set of points.
    translation_matrix = np.array([[1, 0, tx],
                                   [0, 1, ty],
                                   [0, 0, 1]])
    
    #add column with value 1 in each point
    points_homogeneous = np.hstack((points, np.ones((points.shape[0], 1))))
    transformed_points = points_homogeneous @ translation_matrix.T
    #return points ignore last column of value 1
    return transformed_points[:, :2]

def apply_rotation(points, angle, center):
    #Apply rotation to the set of points.
    angle_rad = np.radians(angle)
    rotation_matrix = np.array([[np.cos(angle_rad), -np.sin(angle_rad)],
                                [np.sin(angle_rad),  np.cos(angle_rad)]])
    #Subtract the center from each point. This moves the center of rotation to the origin.
    centered_points = points - center
    rotated_points = centered_points @ rotation_matrix.T + center
    return rotated_points

def apply_scaling(points, sx, sy, center):
    #Apply scaling to the set of points.
    scaling_matrix = np.array([[sx, 0],
                               [0, sy]])
    #Subtract the center from each point. This moves the center of rotation to the origin.
    centered_points = points - center
    scaled_points = centered_points @ scaling_matrix.T + center
    return scaled_points

def get_user_input():
    #Get user input for the type of transformation and parameters.
    transformation_type = input("Enter the transformation type (translation, rotation, scaling): ").strip().lower()
    if transformation_type == "translation":
        tx = float(input("Enter the translation in x direction: "))
        ty = float(input("Enter the translation in y direction: "))
        return transformation_type, tx, ty
    elif transformation_type == "rotation":
        angle = float(input("Enter the rotation angle in degrees: "))
        return transformation_type, angle
    elif transformation_type == "scaling":
        sx = float(input("Enter the scaling factor in x direction: "))
        sy = float(input("Enter the scaling factor in y direction: "))
        return transformation_type, sx, sy
    else:
        print("Invalid transformation type")
        return None, None

def apply_transformation(points, transformation_type, *params):
    #Apply the chosen transformation to the set of points.
    if transformation_type == "translation":
        tx, ty = params
        return apply_translation(points, tx, ty)
    elif transformation_type == "rotation":
        angle = params[0]
        center = np.mean(points, axis=0)  # Use the centroid of the polygon as the center of rotation
        return apply_rotation(points, angle, center)
    elif transformation_type == "scaling":
        sx, sy = params
        center = np.mean(points, axis=0)  # Use the centroid of the polygon as the center of scaling
        return apply_scaling(points, sx, sy, center)
    else:
        return points

def draw_polygon(img, points, color=(0, 255, 0)):
    #Draw a polygon on the image given a set of points.
    pts = points.astype(np.int32).reshape((-1, 1, 2))
    cv2.polylines(img, [pts], isClosed=True, color=color, thickness=2)
    cv2.fillPoly(img, [pts], color)

# Read the image
filename = input("Enter the filename of the image you want to process: ")
img = cv2.imread(filename)

if img is None:
    print(f"File {filename} not found.")
else:
    # Create a window and set the mouse callback function
    cv2.namedWindow('Image')
    cv2.setMouseCallback('Image', on_mouse_click)

    # Display the image and wait for user interaction
    cv2.imshow('Image', img)
    print("Left click to select vertices of the polygon.")
    print("Right click to close the polygon.")
    cv2.waitKey(0)

    if vertices:
        # Get user input for the transformation
        transformation_type, *params = get_user_input()

        if transformation_type:
            # Convert vertices list to numpy array
            points = np.array(vertices, dtype=np.float32)

            # Apply the transformation
            transformed_points = apply_transformation(points, transformation_type, *params)

            # Draw the transformed polygon on the image
            img_transformed = img.copy()
            draw_polygon(img_transformed, transformed_points)

            # Display the transformed image
            cv2.imshow('Transformed Image', img_transformed)
            cv2.waitKey(0)

    cv2.destroyAllWindows()
```
